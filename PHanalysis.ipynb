```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint
import os

# Raw data file
source_file = ("Resources/MovieData_clean.csv")
df = pd.read_csv(source_file)
df.head()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df["Budget"], df["Revenue"], 
            marker="o", 
            facecolors="steelblue", 
            edgecolors="black"
           )
plt.title("Budget vs Revenue)")
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(df["Budget"], df["Revenue"])
regress_values = df["Budget"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.figure(figsize=(15,10))
plt.scatter(df["Budget"], df["Revenue"], 
            marker="o", 
            facecolors="steelblue", 
            edgecolors="black"
           )
plt.plot(df["Budget"],
         regress_values,
         "r-",
         color = "firebrick"
        )
plt.annotate(line_eq,(0, 2000000000),
             fontsize=15,
             color="firebrick"
            )
plt.title("Movie Budget vs Revenue")
plt.xlabel("Budget")
plt.ylabel("Revenue")
print(f"The correlation coefficient is: {rvalue}")
plt.grid(True)
plt.show()
```
